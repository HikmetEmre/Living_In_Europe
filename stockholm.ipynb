```python
from urllib.request import urlopen
import requests
from bs4 import BeautifulSoup as bts
import pandas as pd 
import re
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
def getAndParseURL(url):
    result=requests.get(url,headers={"User-Agent":"Chrome/111.0.5563.65"})
    soup=bts(result.text,"html.parser")
    return soup
```

```python
pages=["https://www.housingtarget.com/sweden/housing-rentals/stockholm"]
for page in range(2,199):
    pages.append("https://www.housingtarget.com/sweden/housing-rentals/stockholm/pageindex" +str (page))
    
pages
```

```python
links = []
for page in pages:
    html = getAndParseURL(page)
    for resulthf in html.findAll("a",{"target":"_blank"}):
        links.append("https://www.housingtarget.com" + resulthf.get("href"))

links
```

```python
result = []
for resulthf in links:
    html = getAndParseURL(resulthf)
    
    try:
        location = html.find("h1",{"id":"ad_title"}).text.strip()
    except:
        location = np.nan
    try:
        rent= html.find("div",{"class":"facts-list"}).find(text=re.compile("Price")).findNext().text.replace(",","").replace("SEK","").strip()
    except:
        rent = np.nan
    try:
        rooms = html.find("div",{"class":"facts-list"}).find(text=re.compile("Number")).findNext().text.strip()
    except:
        rooms = np.nan
    try:
        net_area = html.find("div",{"class":"facts-list"}).find(text=re.compile("Size")).findNext().text.replace("m2","").strip()
    except:
        net_area = np.nan
    try:
        prop_type = html.find("div",{"class":"facts-list"}).find(text=re.compile("Housing")).findNext().text.strip()
    except:
        prop_type = np.nan
    time.sleep(2)
    
    result.append([location,rent,rooms,net_area,prop_type])
    
columns=["Location","Rent","Deposit","Rooms","Bathroom","Net Area"]
df = pd.DataFrame.from_records(result, columns=columns)
```
